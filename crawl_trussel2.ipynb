Grab the pages for all PAN letters

```python
import requests as rq
from bs4 import BeautifulSoup

URL = "https://www.trussel2.com/ACD/"

response = rq.get(URL + "acd-s_j.htm")
bs = BeautifulSoup(response.content, "html.parser")

cognate_pages = list(map(lambda a: a["href"], bs.find("map", {"name": "graph-2"}).select("area")))

for i in cognate_pages:
    print(i)
```

Stuff the parsed soup into context so no more requests needed

```python
cognate_soups = []

for cognate_page in cognate_pages:
    cognate_soups.append(BeautifulSoup(rq.get(URL + cognate_page).content, "html.parser"))
```

```python
import re
```

```python
corpus = []
for bs in cognate_soups:
    settables = bs.find_all("table", class_="settable")
    for settable in settables:
        cognate_table = settable\
            .find("table", class_="forms")
        malay_row = cognate_table.find("span", class_="lg", string="Malay")

        if not malay_row: continue
        
        try:
            # Common practice in dictionaries put subscript numbers
            # after homonyms to distinguish them, for my purpose
            # I simply assume the reflex(cognate in modern language) is
            # going to be the same. So I remove subscript
            pan = re.sub(r'[₁₂₃₄₅₆₇₈₉]', "", settable.find("span", class_="key").string[1:])
            malay = malay_row\
                .find_parent().find_parent()\
                .find_next_sibling()\
                .string
            corpus.append({"pan": pan, "malay": malay})
        except:
            print("entry parsing failed")
corpus
```

```python
import pandas as pd
```

```python
corpus_df = pd.DataFrame(corpus)
corpus_df.to_csv("res/pan2malay.csv")
```
